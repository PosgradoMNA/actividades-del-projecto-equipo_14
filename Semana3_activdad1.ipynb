**PARTE 1. Fundamentos de bases de datos **

**Bases de datos**
Las bases de datos son almacenes de información estructurada que nos permiten coleccionar información para posteriormente esta pueda ser gestionada. Una vez que la información es almacenada en estas estructuras, mediante herramientas como el lenguaje SQL, la información es consultada, mezclada, eliminada o actualizada de acuerdo con el propósito que tenga la recopilación de esta. Dentro del trabajo de la ciencia de datos, las bases de datos entran en el proceso de colección de información, misma que puede ser curada dentro o fuera de la misma base de datos para que esta pueda cumplir con un propósito en específico. Los objetos de almacenamiento en una base de datos se llaman tablas, las cuales son conformadas por elementos llamados campos donde cada uno es especificado con tipo de dato que corresponda con la información que estará albergando. Las tablas de las bases de datos cuando van siendo creadas se les va relacionando unas con otras para que al momento de guardar o consultar la información, los resultados cobren sentido para el analista que extrae la información, todo esto se va realizando con el lenguaje SQL. Para que una base de datos pueda operar, requiere lo que se como gestor de bases de datos (DBMS , por sus siglas en inglés database management system), entre los más importantes están los que desarrolla la empresa Oracle.

**Datawarehouses**
Los almacenes de información (Datawarehouse) son grandes concentradores de información, la cual proviene de múltiples fuentes de datos, como sistemas transaccionales, bases de datos. Esta información concentrada se utiliza para el análisis de datos empresarial y es base para la inteligencia de negocios.

```python
import pandas as pd
import numpy as np
```

```python
#Cargar el dataset
data_path = "https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
# ¿Faltan datos ?
df.isnull().values.any()
```

```python
#Test 1 . Metodo de Eliminación valores nulos
df.dropna(subset=['X2', 'X3', 'X4', 'X5', 'Y'], inplace=True)
df
```

```python
#Test 2 . Proceso de media para crear valores faltantes
columnsToSubstitute = ['X6', 'X7', 'X8', 'X9', 'X10', 'X11', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20', 'X21', 'X22', 'X23']
for col in columnsToSubstitute:
  mediana = df[col].median()
  df[col] = df[col].fillna(mediana)

df
```

¿Se eliminaron o reemplazaron datos nulos? ¿Qué se hizo y por qué?

Al haber una cnatidad de datos muy baja , puede ser un metodo bueno el eliminar, al probar la sustitución de la media en valores faltantes se debe emplear cuando existan un mayor numero de datos con nulos , para evitar perder o caer temas donde el modelo no pueda aprender lo suficiente.
